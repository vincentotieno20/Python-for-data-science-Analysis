#### Phase 2 - Group data by school ratings

```python
grouped_data = middle_tn_schools.groupby('school_rating')
reduced_lunch_stats = grouped_data['reduced_lunch'].describe()
reduced_lunch_stats
```

#### Phase 3 – Correlation analysis

```python
correlation_matrix = middle_tn_schools[['reduced_lunch', 'school_rating']].corr()
correlation_matrix
```

#### Phase 4 – Scatter Plot 

```python
## Set the style for the plot
sns.set(style="whitegrid")

## Create a figure and axis
plt.figure(figsize=(10, 6))

## Create a scatter plot
sns.scatterplot(data=middle_tn_schools, x='reduced_lunch', y='school_rating')

## Add a trend line
sns.regplot(data=middle_tn_schools, x='reduced_lunch', y='school_rating', scatter=False, color='yellow')
## Add labels
plt.xlabel('Reduced Lunch (%)')
plt.ylabel('School Rating')
plt.title('Scatter Plot of School Rating vs. Reduced Lunch')
## Desplay the plot
plt.show()
```
